```python
import pandas as pd
thing = pd.ExcelFile('TheData.xlsx')
thing.sheet_names
```

```python
average = thing.parse('Average_House')
average.fillna(0, inplace =True)
average
```

```python
interestRate = pd.read_excel('Book1.xlsx')
interestRate
```

```python
results = []
for x in average['observation_date']:
    try: 
        results.append(interestRate[interestRate['year'] == x.year]['interest'].values[0])
    except:
        results.append(0)
average['inflation'] = results
results = []
for x in average['observation_date']:
    try: 
        results.append(interestRate[interestRate['year'] == x.year]['compound intrest'].values[0] +1)
    except:
        results.append(0)
average['compound_Rate'] = results
average['Inflation_MW'] = [x * y for x,y in zip(average['Average_MW'],average['compound_Rate'])]
average['Inflation_NE'] = [x * y for x,y in zip(average['Average_NE'],average['compound_Rate'])]
average['Inflation_PS'] = [x * y for x,y in zip(average['Average_PS'],average['compound_Rate'])]
average['Inflation_PW'] = [x * y for x,y in zip(average['Average_PW'],average['compound_Rate'])]
```

```python
median = thing.parse('Median_House')
median.fillna(0, inplace =True)
median.drop(columns = 'Unnamed: 8', inplace = True)
results = []
for x in median['date']:
    try: 
        results.append(interestRate[interestRate['year'] == x.year]['interest'].values[0])
    except:
        results.append(0)
median['inflation'] = results
results = []
for x in median['date']:
    try: 
        results.append(interestRate[interestRate['year'] == x.year]['compound intrest'].values[0] +1)
    except:
        results.append(0)
median['compound_Rate'] = results
median['Inflation_MW'] = [x * y for x,y in zip(median['Median_MW'],median['compound_Rate'])]
median['Inflation_NE'] = [x * y for x,y in zip(median['Median_NE'],median['compound_Rate'])]
median['Inflation_PS'] = [x * y for x,y in zip(median['Median_PS'],median['compound_Rate'])]
median['Inflation_PW'] = [x * y for x,y in zip(median['Median_PW'],median['compound_Rate'])]
median
```

```python
cleanedData= pd.ExcelWriter('Cleaned_Data.xlsx')
average.to_excel(cleanedData, sheet_name= 'AveragePrice', index = False)
median.to_excel(cleanedData, sheet_name= 'MedianPrice', index = False)
thing.parse('Regions').to_excel(cleanedData, sheet_name = 'Regions')
thing.parse('NewHousesSold(Thousand)').to_excel(cleanedData, sheet_name = 'NewHousesSold(Thousand)', index = False)
thing.parse('NewHouseBuilt_Thousand').to_excel(cleanedData, sheet_name = 'NewHouseBuilt_Thousand', index = False)
cleanedData.close()
```
