```python
file_path3 = r"C:\Users\Ajith\OneDrive\Desktop\CountyIncBreakdown.xlsx"
dfStates = pd.read_excel(file_path3)
dfStates = dfStates.dropna(how = 'all')
dfStates.head()
```

```python
state_to_region = {}
for region, states in us_regions.items():
    for state in states:
        state_to_region[state] = region

# Map each state to its region 
dfStates['Region'] = dfStates['State'].map(state_to_region)
dfStates.head()
```
